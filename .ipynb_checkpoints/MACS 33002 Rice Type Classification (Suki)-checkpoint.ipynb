## **Part-2: Data Preprocessing**

In this section, we focus on making the data ready to use for ML models. It includes:

- dropping unnecessary fields
- filling in missing values
- checking outliers
- rearranging data

### dropping unnecessary fields

```python
# We drop the 'id' column because it isn't useful in the analysis or prediction. We also drop one of ConvexArea or 
# EquivDiameter because it has the almost the same correlation values across all predictors and target variables.
rice_data_df = rice_data_df.drop(['id', 'ConvexArea'], axis=1)
```

### filling in missing values

```python
# We can see there are no missing elements in the data dataframe. The data is clean and doesn't need to replace. 
rice_data_df.isnull().sum()
```

### checking outliers

```python
# draw box plot to check outliers
fig, axes = plt.subplots(5,2, figsize=(20,25))
i=0
for j in range(0,5):
    for k in range(0,2):
        sns.boxplot(ax=axes[j,k], data=rice_data_df, x=rice_data_df.columns[i], palette='Paired')
        i=i+1
```

Because there is no significant ourlier, we decide not to remove any sample.
However, since some variables take values between 0 and 1, while others take values greater than 1000. We would like to standarize features that take values greater than 1 without distorting the differences in the range of the values.

### Rearranging data by standarrization

```python
scaler = MinMaxScaler(feature_range=(0,1))
trans_cols = ['MajorAxisLength', 'MinorAxisLength', 'ConvexArea', 'EquivDiameter', 'Perimeter', 'AspectRation']

for col in trans_cols:
    rice_data_df[col] = scaler.fit_transform(rice_data_df[[col]].values)
    
rice_data_df.head()
```

## **Part-3.1: Algorithm 1 (DT classifier & Random Forest)**

### DT classifier

```python
feature_cols = ["MajorAxisLength", "MinorAxisLength", "Eccentricity", 'EquivDiameter', "Extent", "Perimeter", "Roundness", "AspectRation"]
X = rice_data_df[feature_cols]
y = rice_data_df.Class
```

```python
X.columns, y
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size=0.75, test_size=0.25) 
```

```python
# Create Decision Tree classifer object
DT_clf = DecisionTreeClassifier(random_state = 0, criterion = 'entropy', max_depth=3, min_samples_split=10, min_samples_leaf=3)
# We use max_depth = 3, min_samples_split = 10 and min_samples_leaf = 3 in order to make the visualization easier 
# to read and to limit pre/post - pruning

# Train Decision Tree Classifer
DT_clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = DT_clf.predict(X_test)
```

```python
#Score of the model
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Visualization
dot_data = tree.export_graphviz(DT_clf, out_file=None, 
                                feature_names=X.columns,  
                                class_names=["Gonen", "Jasmine"],  
                                filled=True, rounded=True,  
                                special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

Looking at the graph above, "MinorAxisLength" is the most frequent and important detergent of classifing the rice type in DT model.
The model also use "AspectRation", "Eccentricity", and "Roundness" to classify the rice type.

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
RF_clf = RandomForestClassifier(random_state=0, criterion = 'entropy', max_depth=3, min_samples_split=10, min_samples_leaf=3)
RF_clf.fit(X_train, y_train)
```

```python
#Score of the model
print("Accuracy:",np.round(RF_clf.score(X_test, y_test), decimals=4))
```

```python
#Visualization
dot_data_2 = tree.export_graphviz(RF_clf.estimators_[0], out_file=None, 
                                  filled=True, rounded=True,  
                                  special_characters=True)  
graph_2 = graphviz.Source(dot_data_2)  
graph_2
```

### Comparison of the accuracy of DT and RF

The accuracy scores of our random forest is higher than the score of our decision tree model (0.9875 > 0.9839).
This result represents that random forest made an improvement of robustness.

## **Part-3.2: Algorithm 2 (Logistic Regression)**

```python
rice = rice_data_df
# rename the original dataframe "rice_data_df" to "rice" for easier running the cells below
```

```python
# We use the IQR method to remove outliers and we have to balance the data after removing outliers because logsitic regression is sensitive to imbalanced data and outliers. Imbalance data will give us a biased result and outliers can affect the model's performance.

# IQR method to remove outliers

Q1 = rice.quantile(0.25)
Q3 = rice.quantile(0.75)
IQR = Q3 - Q1
rice = rice[~((rice < (Q1 - 1.5 * IQR)) |(rice > (Q3 + 1.5 * IQR))).any(axis=1)]
rice.shape
```

```python
X = rice.drop(columns=['id', 'Area', 'ConvexArea', 'Class'], axis=1)
y = rice['Class']
```

```python
# rice data is not balanced, we have to balance the data
# we can use SMOTE method to balance the data

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
print('Original dataset shape %s' % Counter(y))
print('Resampled dataset shape %s' % Counter(y_resampled))
```

```python
X_resampled.shape
```

```python
y_resampled.shape
```

```python
#the data is balanced now
y_resampled.value_counts()
```

```python
# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.25, random_state=42) 
```

### Create a Logistic Regression for rice with L1 penalty

```python
#add training data to the model

clf = LogisticRegression(penalty='l1', solver='liblinear', random_state=42, max_iter=100000).fit(X_train, y_train)

#predict the test data
y_pred = clf.predict(X_test)
#check the accuracy
accuracy_score(y_test, y_pred)
#check the confusion matrix
confusion_matrix(y_test, y_pred)
#check the classification report
print(classification_report(y_test, y_pred))
#check the roc_auc_score
roc_auc_score(y_test, y_pred)
#check the roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
#check the precision_recall_curve
precision, recall, thresholds = precision_recall_curve(y_test, y_pred)
#check the auc
auc(fpr, tpr)
#check the f1_score
f1_score(y_test, y_pred)
#check the precision_score
precision_score(y_test, y_pred)
#check the recall_score
recall_score(y_test, y_pred)
#check the average_precision_score
average_precision_score(y_test, y_pred)
```

```python
# check the top coef to see if the classifier is built with reasonable features
feat_importances = pd.Series(clf.coef_[0], index=X_resampled.columns)
feat_importances.nlargest(8).plot(kind='barh')
plt.show()
```

### Model Explaination and Evaluation 

- For both the decision tree and KNN classifier we use the default 75% training and 25% test. We trained both models with all the feature variables minus 'id' and 'EquivDiameter' and the target 'Class' variable.

- Using the score function, we can check the accuracy of each model. 
- For the decision tree, it scored 0.9839
- For the random forest, it scored 0.9875
- For the knn classifer at K = 3, 5, 10 the corresponding accuracy scores are: 0.9804266549373213, 0.9795469540356279, and 0.977787552232241. 
- For every value of K, the knn classifier model does not perform better than the decision tree. In addition at there seems to be a negative correlation between k and accuracy score with the knn classifer model.

- In conclusion, the decision tree is more accurate and faster than the knn classifier model for the binary classification of the rice data types. (It is faster because knn classifer has to scan the whole dataset to make predictions, while decision trees do not.)

It is very difficult to set a suitable K for a given training dataset. The errors of the Knn classifer are largely due to the fact that it considers the entire set of n points for every execution, as a result at certain K levels, errors can be made because of some points outside its known cluster/outliers which lead to faulty predictions. 

From the confusion matrixes above, the Knn classifiers in total make more errors than the decison tree. This makes sense because the decision tree model is more accurate than the Knn classifer models. The Knn classifer models make 89, 93 and 101 errors for K values of 3, 5 and 10 respectively. The decision tree made 59 errors in comparison. Also, although the Knn classifier made more total errors, they made less errors when predicting Gonen when the true value was Jasmine in relation to the decision tree model.
